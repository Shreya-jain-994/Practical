```python
import docx2txt
job_description = docx2txt.process('/content/Job Description for Data Analyst.docx')
resume_of_HR = docx2txt.process('/content/HR Resume.docx')
resume_of_analyst = docx2txt.process('/content/Entry Level Data Analyst Resume.docx')
```

```python
match_percentage_hr = [job_description,resume_of_HR]
match_percentage_analyst = [job_description,resume_of_analyst]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
```

```python
matrix1 = cv.fit_transform(match_percentage_hr)
matrix2 = cv.fit_transform(match_percentage_analyst)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_matrix_1 = cosine_similarity(matrix1)
similarity_matrix_2= cosine_similarity(matrix2)
```

```python

print(similarity_matrix_1)
```

```python
print(similarity_matrix_2)
```

```python
print('The resume of HR candidate matches by ' + str(similarity_matrix_1))
print('The resume of Analyst candidate matches by ' + str(similarity_matrix_2))
```
